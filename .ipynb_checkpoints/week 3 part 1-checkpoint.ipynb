# Neighborhoods in Toronto

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find('tbody')
rows = table.select('tr')
row = [r.get_text() for r in rows]
```

```python
wiki = pd.DataFrame(row)
wiki = wiki[0].str.split('\n', expand=True)
wiki = wiki.rename(columns=wiki.iloc[0])
wiki = wiki.drop(wiki.index[0])
wiki.head()
```

```python
wiki.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)
```

```python
wiki = wiki[wiki.Borough != 'Not assigned']
wiki.head()
```

```python
wiki = wiki.groupby(['PostalCode', 'Borough'], sort = False).agg(','.join)
wiki.reset_index(inplace = True)
wiki.head()
```

```python
wiki = wiki.replace("Not assigned", "Queen's Park")
wiki.head()
```

```python
wiki.shape
```

```python
wiki.to_csv('toronto.csv', index = False)
```
